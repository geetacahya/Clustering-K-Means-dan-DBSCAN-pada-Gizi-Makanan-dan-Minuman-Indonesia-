### 2.2 Metode Analisis Data 1: [K-Means]



K-Means adalah salah satu algoritma unsupervised learning yang paling populer untuk analisis clustering. Algoritma ini bertujuan untuk membagi sekumpulan data ke dalam sejumlah klaster berdasarkan kemiripan (similarity) antar data. K-Means bekerja dengan cara meminimalkan jarak antar data ke pusat klaster (centroid) terdekatnya, sehingga setiap data akan dikelompokkan ke klaster yang memiliki centroid paling dekat.

Algoritma K-Means pertama kali diperkenalkan oleh MacQueen pada tahun 1967 dan sejak itu digunakan secara luas dalam berbagai bidang, termasuk pengelompokan data pelanggan, pengolahan citra, pengelompokan dokumen, serta analisis pola konsumsi makanan dan minuman berdasarkan kandungan gizinya.

Rumus utama yang digunakan dalam algoritma K-Means adalah untuk menghitung total within-cluster sum of squares (WCSS), yang menjadi fungsi objektif untuk diminimalkan:
\begin{equation}
\text{WCSS} = \sum_{i=1}^{k} \sum_{x_j \in C_i} \| x_j - \mu_i \|^2
\end{equation}
 dengan:
- $k$ : Jumlah klaster yang ditentukan.
- $C_i$ : Klaster ke-$i$.
- $x_j$ : Titik data ke-$j$ yang termasuk dalam klaster $C_i$.
- $\mu_i$ : Titik pusat (centroid) dari klaster $C_i$.
- $\| x_j - \mu_i \|^2$ : Jarak Euclidean kuadrat antara titik data $x_j$ dan centroid $\mu_i$.

Metode K-Means dipilih karena mampu mengelompokkan data secara efisien berdasarkan kemiripan nilai numerik, seperti kandungan kalori, protein, lemak, dan karbohidrat pada makanan dan minuman. Dengan pendekatan berbasis centroid, metode ini memungkinkan identifikasi kelompok makanan yang memiliki karakteristik gizi yang mirip, sehingga memudahkan dalam interpretasi dan analisis pola konsumsi. Selain itu, K-Means juga cocok diterapkan pada dataset berskala besar dan menghasilkan klaster yang jelas serta terstruktur, yang mendukung kebutuhan analisis gizi secara sistematis.

#### Standarisasi Data (Scaling)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(gizi)
df_scaled = pd.DataFrame(df_scaled, columns=gizi.columns)
df_scaled.head()
```

Kode tersebut melakukan proses standarisasi data menggunakan StandardScaler, yang mengubah setiap fitur gizi (kalori, protein, lemak, karbohidrat) agar memiliki nilai rata-rata 0 dan standar deviasi 1. #### Jumlah Klaster Optimal

1. Elbow Method

```python
from sklearn.cluster import KMeans

wcss = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

# Plot Elbow
plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Jumlah Klaster (K)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.xticks(K_range)
plt.grid(True)
plt.show()
```

Grafik di atas menunjukkan metode elbow untuk menentukan jumlah klaster yang optimal dalam analisis klaster. Dengan mengambil k=4, terlihat bahwa pada titik tersebut, penurunan WCSS (Within-Cluster Sum of Squares) mulai melambat secara signifikan, mengindikasikan bahwa tambahan klaster setelah k=4 tidak memberikan peningkatan yang signifikan dalam pengurangan jarak antar titik dalam klaster. Ini menandakan bahwa k=4 adalah pilihan yang baik untuk analisis klaster.

2. Silhouette Score

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(df_scaled)
    score = silhouette_score(df_scaled, labels)
    silhouette_scores.append(score)

# Plot Silhouette Score
plt.figure(figsize=(8, 5))
plt.plot(K_range, silhouette_scores, marker='o', color='orange')
plt.title('Silhouette Score')
plt.xlabel('Jumlah Klaster (K)')
plt.ylabel('Silhouette Score')
plt.xticks(K_range)
plt.grid(True)
plt.show()
```

Grafik tersebut menunjukkan hubungan antara jumlah klaster (K) dan nilai Silhouette Score. Dari grafik, terlihat bahwa nilai Silhouette Score tertinggi terjadi pada K=4, dengan skor mencapai lebih dari 0.5. Ini menunjukkan bahwa dengan 4 klaster, data terpisah dengan baik satu sama lain. Setelah K=4, nilai Silhouette Score cenderung menurun, menunjukkan bahwa penambahan jumlah klaster selanjutnya mungkin tidak memberikan pembagian yang lebih baik.

### 2.3 Metode Analisis Data 2: [DBSCAN]

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) merupakan salah satu algoritma unsupervised learning yang digunakan untuk mengelompokkan data berdasarkan kepadatan titik data. Algoritma ini dikembangkan oleh Ester et al. pada tahun 1996 dan sangat efektif dalam mengidentifikasi klaster dengan bentuk arbitrer serta dalam mendeteksi outlier atau noise secara otomatis.

Berbeda dengan metode berbasis centroid, DBSCAN mendefinisikan klaster sebagai area yang padat dengan data, dan memisahkannya dari area yang jarang data. Dengan demikian, DBSCAN dapat membentuk klaster tanpa perlu menentukan jumlah klaster di awal, yang menjadikannya fleksibel terhadap distribusi data yang tidak beraturan.

Dua parameter utama dalam DBSCAN:
1. $ \varepsilon $ (epsilon): Jarak maksimum antar titik yang masih dianggap sebagai tetangga dalam radius yang sama.
2. MinPts: Jumlah minimum titik dalam radius Îµ untuk membentuk area padat yang disebut core point.

Jenis Titik dalam DBSCAN:
1. Core Point: Titik yang memiliki paling sedikit MinPts tetangga dalam radius Îµ.
2. Border Point: Titik yang tidak cukup padat untuk menjadi core point, tetapi berada dalam jangkauan Îµ dari core point.
3. Noise Point: Titik yang tidak termasuk dalam klaster mana pun.

Rumus utama dalam DBSCAN berkaitan dengan jarak Euclidean antar titik data untuk menentukan kepadatan lokal:
\begin{equation}
d(p, q) = \sqrt{\sum_{i=1}^{n}(p_i - q_i)^2}
\end{equation}
- $d(p, q)$ : jarak Euclidean antara titik data $p$ dan $q$
- $p_i, q_i$ : nilai dari fitur ke-$i$ pada titik $p$ dan $q$
- $n$ : jumlah dimensi atau fitur

Sebuah titik $p$ dikatakan sebagai core point jika::

\begin{equation}
|\{q \in D \mid d(p, q) \leq \varepsilon\}| \geq \text{MinPts}
\end{equation}
dengan:
- $D$ : himpunan semua data
- $\varepsilon$ : radius maksimum untuk neighborhood
- $\text{MinPts}$ : jumlah minimum titik untuk dianggap sebagai cluster padat

Metode DBSCAN dipilih karena mampu mengelompokkan data berdasarkan kepadatan tanpa perlu menentukan jumlah klaster di awal. Metode ini juga efektif dalam mengenali bentuk klaster yang tidak beraturan serta secara otomatis mengidentifikasi data yang menyimpang atau tidak termasuk ke dalam kelompok mana pun (noise). Dengan kemampuan tersebut, DBSCAN cocok digunakan untuk data kandungan gizi yang mungkin memiliki distribusi kompleks dan tidak linier, sehingga dapat menghasilkan pengelompokan yang lebih alami sesuai struktur data sebenarnya.

#### Standarisasi Data (Scaling)

#### Penentuan Parameter DBSCAN

```python
!pip install kneed
import numpy as np
from kneed import KneeLocator
from sklearn.neighbors import NearestNeighbors

min_samples = 5
neighbors = NearestNeighbors(n_neighbors=min_samples)
neighbors_fit = neighbors.fit(data_scaled)
distances, indices = neighbors_fit.kneighbors(data_scaled)

k_distances = np.sort(distances[:, -1])

kneedle = KneeLocator(range(len(k_distances)), k_distances, curve="convex", direction="increasing")

plt.figure(figsize=(8, 4))
plt.plot(k_distances, label='K-Distance')
plt.axvline(x=kneedle.knee, color='r', linestyle='--', label=f'Elbow at {round(k_distances[kneedle.knee], 2)}')
plt.title(f'K-Distance Graph (MinPts = {min_samples})')
plt.xlabel('Data Point Index')
plt.ylabel(f'Distance to {min_samples}th Nearest Neighbor')
plt.ylim(0, 2)
plt.grid(True)
plt.legend()
plt.show()

print(f"Elbow ditemukan di eps â‰ˆ {round(k_distances[kneedle.knee], 2)}")
```

Gambar yang ditampilkan adalah grafik K-Distance yang digunakan untuk menentukan parameter pada algoritma DBSCAN. Grafik ini menunjukkan jarak ke tetangga kelima terdekat (K=5) untuk setiap titik data, di mana sumbu X merepresentasikan indeks titik data dan sumbu Y menunjukkan jarak. Titik "elbow" pada grafik, yang terlihat di sekitar jarak 1.02 (dibulatkan 1.0), membantu dalam menentukan epsilon (Îµ) sebagai jarak ambang untuk mengklasifikasikan titik sebagai "core point" dalam DBSCAN. Pemilihan nilai Îµ yang tepat penting untuk mengidentifikasi kluster dengan baik, memastikan bahwa titik-titik dalam jarak tersebut saling terhubung dan membentuk kelompok yang relevan.

### 3.1 K-Means

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Atur ukuran marker dan palet warna
palette = sns.color_palette("Set2", 4)
markersize = 80

# Kombinasi fitur yang akan divisualisasikan
feature_pairs = [
    ('calories', 'proteins'),
    ('calories', 'fat'),
    ('calories', 'carbohydrate'),
    ('proteins', 'fat'),
    ('proteins', 'carbohydrate'),
    ('fat', 'carbohydrate'),
]

# Plot semua kombinasi fitur
for x_feat, y_feat in feature_pairs:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=x_feat, y=y_feat, hue='cluster', palette=palette, s=markersize)
    plt.title(f'Visualisasi Klaster (K-Means): {x_feat.capitalize()} vs {y_feat.capitalize()}')
    plt.xlabel(x_feat.capitalize())
    plt.ylabel(y_feat.capitalize())
    plt.legend(title='Klaster')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

Beberapa titik dengan nilai carbohydrate yang sangat tinggi menunjukkan keberadaan outlier dalam data. Hal ini mengindikasikan bahwa model K-Means sensitif terhadap nilai-nilai ekstrem, sehingga titik-titik tersebut tetap dimasukkan ke dalam salah satu klaster meskipun secara karakteristik gizi sangat berbeda dari sebagian besar data. Kondisi ini dapat memengaruhi posisi centroid dan struktur klaster secara keseluruhan, sehingga penting untuk mempertimbangkan efek outlier dalam interpretasi hasil klastering.

#### Metrik Evaluasi Model K-Means

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Gunakan hasil label dari K-Means
labels_kmeans = df['cluster']

# Hitung metrik evaluasi
silhouette = silhouette_score(df_scaled, labels_kmeans)
davies_bouldin = davies_bouldin_score(df_scaled, labels_kmeans)
calinski_harabasz = calinski_harabasz_score(df_scaled, labels_kmeans)

# Tampilkan hasil
print("Metrik Evaluasi K-Means:")
print(f"Silhouette Score       : {silhouette:.4f}  (Semakin tinggi semakin baik)")
print(f"Davies-Bouldin Index   : {davies_bouldin:.4f}  (Semakin rendah semakin baik)")
print(f"Calinski-Harabasz Score: {calinski_harabasz:.4f}  (Semakin tinggi semakin baik)")
```

Hasil evaluasi model K-Means menunjukkan bahwa nilai Silhouette Score sebesar 0.5129 termasuk dalam kategori sedang hingga baik, yang menandakan bahwa pemisahan antar klaster cukup jelas dan data dalam klaster cukup kompak. Davies-Bouldin Index bernilai 0.8728, yang relatif rendah, mengindikasikan bahwa klaster yang terbentuk memiliki jarak yang baik satu sama lain dengan tingkat tumpang tindih yang minim. Sementara itu, Calinski-Harabasz Score sebesar 787.6491 menunjukkan nilai yang cukup tinggi, menandakan bahwa rasio antara dispersi antar klaster dan dalam klaster cukup optimal. Secara keseluruhan, ketiga metrik ini mengindikasikan bahwa model K-Means mampu membentuk klaster yang cukup baik dan dapat merepresentasikan struktur alami dari data kandungan gizi makanan dan minuman.


### 3.2 DBSCAN

#### Hasil Visualisasi Model DBSCAN

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter hanya data yang bukan noise
data_klaster = data[data['cluster'] != -1]

# Atur ukuran marker dan palet warna sesuai jumlah klaster non-noise
n_clusters = data_klaster['cluster'].nunique()
palette = sns.color_palette("Set2", n_colors=n_clusters)
markersize = 80

# Kombinasi fitur yang akan divisualisasikan
feature_pairs = [
    ('calories', 'proteins'),
    ('calories', 'fat'),
    ('calories', 'carbohydrate'),
    ('proteins', 'fat'),
    ('proteins', 'carbohydrate'),
    ('fat', 'carbohydrate'),
]

# Plot semua kombinasi fitur tanpa noise
for x_feat, y_feat in feature_pairs:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=data_klaster, x=x_feat, y=y_feat, hue='cluster', palette=palette, s=markersize)
    plt.title(f'Visualisasi Klaster (Tanpa Noise): {x_feat.capitalize()} vs {y_feat.capitalize()}')
    plt.xlabel(x_feat.capitalize())
    plt.ylabel(y_feat.capitalize())
    plt.legend(title='Klaster')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

Visualisasi di atas menunjukkan klaster dari data makanan berdasarkan beberapa parameter gizi. Warna yang digunakan untuk klaster menunjukkan dua kelompok: klaster 0 (diberi warna hijau) dan klaster 1 (diberi warna oranye). Setiap grafik mengeksplorasi hubungan antara dua parameter gizi (seperti Fat vs Carbohydrate, Proteins vs Carbohydrate, dan lainnya) untuk menilai distribusi dan keterkaitan antar variabel. Visualisasi ini sudah tidak menampilkan outlier (data dengan label -1), karena data noise telah dihilangkan sebelumnya. Hal ini mengindikasikan bahwa DBSCAN berhasil mengidentifikasi dan mengeluarkan outlier dari proses klastering, sehingga hanya data dengan kemiripan kepadatan yang cukup dimasukkan ke dalam klaster. Dengan kata lain, DBSCAN tidak hanya mampu membentuk klaster yang representatif terhadap distribusi alami data, tetapi juga secara efektif menangani data-data ekstrem atau menyimpang yang berpotensi mengganggu hasil pengelompokan.

#### Metrik Evaluasi Model DBSCAN

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Ambil label hasil DBSCAN
labels = data['cluster']

# Filter data yang bukan noise (-1)
mask = labels != -1
data_clustered = data_scaled[mask]
labels_clustered = labels[mask]

# Hitung metrik evaluasi hanya untuk data yang termasuk dalam klaster
if len(set(labels_clustered)) >= 2:  # Minimal 2 klaster agar metrik valid
    silhouette = silhouette_score(data_clustered, labels_clustered)
    db_index = davies_bouldin_score(data_clustered, labels_clustered)
    ch_score = calinski_harabasz_score(data_clustered, labels_clustered)

    print("Metrik Evaluasi DBSCAN:")
    print(f"Silhouette Score       : {silhouette:.4f}  (Semakin tinggi semakin baik)")
    print(f"Davies-Bouldin Index   : {db_index:.4f}  (Semakin rendah semakin baik)")
    print(f"Calinski-Harabasz Score: {ch_score:.4f}  (Semakin tinggi semakin baik)")
```

Hasil evaluasi model DBSCAN menunjukkan bahwa nilai Silhouette Score sebesar 0.7243 tergolong tinggi, yang berarti klaster yang terbentuk memiliki struktur yang sangat baik, dengan data yang cenderung saling mendekat dalam klaster dan berjauhan antar klaster. Nilai Davies-Bouldin Index sebesar 0.2652 menunjukkan tingkat tumpang tindih antar klaster yang sangat rendah, yang mengindikasikan bahwa klaster-klaster tersebut cukup terpisah dan padat secara internal. Sementara itu, Calinski-Harabasz Score** sebesar 245.4261 juga cukup tinggi, menunjukkan rasio yang baik antara dispersi antar klaster dan dalam klaster. Secara keseluruhan, metrik evaluasi ini menunjukkan bahwa DBSCAN membentuk klaster yang jelas, kompak, dan memiliki pemisahan yang efektif berdasarkan pola kepadatan data kandungan gizi.

3.3 Pemilihan Model Terbaik

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from tabulate import tabulate

# ================== EVALUASI K-MEANS ==================
labels_kmeans = df['cluster']
silhouette_kmeans = silhouette_score(df_scaled, labels_kmeans)
db_index_kmeans = davies_bouldin_score(df_scaled, labels_kmeans)
ch_score_kmeans = calinski_harabasz_score(df_scaled, labels_kmeans)

# ================== EVALUASI DBSCAN ==================
labels_dbscan = data['cluster']
mask_dbscan = labels_dbscan != -1
data_dbscan_filtered = data_scaled[mask_dbscan]
labels_dbscan_filtered = labels_dbscan[mask_dbscan]

# Evaluasi DBSCAN hanya jika jumlah klaster minimal 2
if len(set(labels_dbscan_filtered)) >= 2:
    silhouette_dbscan = silhouette_score(data_dbscan_filtered, labels_dbscan_filtered)
    db_index_dbscan = davies_bouldin_score(data_dbscan_filtered, labels_dbscan_filtered)
    ch_score_dbscan = calinski_harabasz_score(data_dbscan_filtered, labels_dbscan_filtered)
else:
    silhouette_dbscan = db_index_dbscan = ch_score_dbscan = None

# ================== TABEL PERBANDINGAN ==================
table = [
    ["Silhouette Score", f"{silhouette_kmeans:.4f}", f"{silhouette_dbscan:.4f}"],
    ["Davies-Bouldin Index", f"{db_index_kmeans:.4f}", f"{db_index_dbscan:.4f}"],
    ["Calinski-Harabasz Score", f"{ch_score_kmeans:.4f}", f"{ch_score_dbscan:.4f}"],
]

headers = ["Metrik Evaluasi", "K-Means", "DBSCAN"]
print(tabulate(table, headers=headers, tablefmt="grid"))
```

Berdasarkan hasil evaluasi yang ditampilkan pada tabel, terlihat bahwa model DBSCAN menunjukkan performa yang lebih baik dibandingkan K-Means dalam mengelompokkan data makanan dan minuman berdasarkan kandungan gizinya. Hal ini terlihat dari nilai Silhouette Score DBSCAN yang lebih tinggi, yaitu 0.7243, dibandingkan dengan K-Means yang hanya sebesar 0.5129. Skor ini menunjukkan bahwa klaster yang dibentuk oleh DBSCAN lebih kompak dan lebih terpisah satu sama lain, mencerminkan struktur klaster yang lebih jelas.

Selain itu, Davies-Bouldin Index DBSCAN bernilai lebih rendah (0.2652) dibandingkan K-Means (0.8728), yang mengindikasikan bahwa klaster yang dihasilkan oleh DBSCAN memiliki tingkat tumpang tindih yang lebih kecil antar klaster. Semakin rendah nilai indeks ini, semakin baik pemisahan antar klaster dan semakin padat data dalam masing-masing klaster.

Namun, pada Calinski-Harabasz Score, K-Means menghasilkan nilai yang jauh lebih tinggi (787.649) dibandingkan DBSCAN (245.426), yang berarti secara rasio antara dispersi antar dan dalam klaster, K-Means tampak lebih stabil. Meskipun demikian, skor ini bisa terpengaruh oleh jumlah data dalam klaster, dan karena DBSCAN mengecualikan noise dari proses penghitungan, skor tersebut tidak sepenuhnya mencerminkan kualitas model secara keseluruhan.

Secara keseluruhan, dari ketiga metrik evaluasi yang digunakan, DBSCAN dinilai sebagai model yang lebih optimal dalam menangkap struktur alami data gizi makanan dan minuman, terutama karena kemampuannya dalam mengidentifikasi klaster dengan kepadatan tinggi dan memisahkan outlier secara otomatis dari proses pengelompokan.

3.4 Kelebihan dan Keterbatasan Model DBSCAN

Model DBSCAN memiliki kelebihan utama dalam kemampuannya mengelompokkan data berdasarkan kepadatan tanpa harus menentukan jumlah klaster di awal. Hal ini sangat berguna ketika struktur klaster dalam data tidak diketahui secara pasti atau tidak berbentuk simetris. Dalam analisis ini, DBSCAN berhasil membentuk klaster yang jelas dan memisahkan 14 data sebagai outlier, menunjukkan keunggulannya dalam mendeteksi data yang menyimpang dari pola umum. Nilai Silhouette Score yang tinggi dan Davies-Bouldin Index yang rendah juga mendukung bahwa klaster yang terbentuk padat secara internal dan terpisah dengan baik satu sama lain.

Namun, model ini juga memiliki keterbatasan. DBSCAN sensitif terhadap pemilihan parameter eps dan min_samples. Jika parameter ini tidak diatur dengan tepat, model dapat menghasilkan terlalu sedikit atau terlalu banyak klaster, atau bahkan mengklasifikasikan sebagian besar data sebagai noise. Selain itu, DBSCAN kurang efektif ketika data memiliki variasi kepadatan yang tinggi antar klaster, karena semua klaster diasumsikan memiliki tingkat kepadatan yang serupa. Dalam konteks ini, meskipun DBSCAN memberikan hasil klaster yang baik, jumlah klaster yang terbentuk relatif sedikit dibandingkan kompleksitas data, sehingga mungkin belum sepenuhnya menggambarkan keragaman kandungan gizi dari seluruh jenis makanan dan minuman yang dianalisis.

Dengan demikian, DBSCAN sangat cocok digunakan ketika pola data menunjukkan kepadatan lokal yang jelas dan ada potensi keberadaan outlier, namun perlu kehati-hatian dalam pemilihan parameter agar model dapat bekerja secara optimal.
